## Set up

```python
# load libraries
import os
import urllib.request
import pandas as pd
import plotnine as p9
%matplotlib inline
import warnings
import matplotlib.pyplot as plt
```

```python
# get csv
urllib.request.urlretrieve("https://github.com/RamiKrispin/coronavirus/raw/master/csv/coronavirus.csv",
                           "coronavirus.csv")
```

```python
# pull data into notebook and save as the variable 'covid'
covid = pd.read_csv("coronavirus.csv")
```

```python
# take a peek at the dataset
# take note of variable names and data types
covid.head()
```

## Practice problems

### Practice problem 1:

How many countries are represented by this dataset? Which country is most frequently in the data?

```python
covid.country.describe()
```

**Answer**: 192 countries, China occurs the most often

### Practice problem 2:

Subset the dataset to only show data for Afghanistan.

```python
covid[(covid.country == "Afghanistan")]
```

### Practice problem 3:

Show the data for confirmed cases of COVID-19 in the United States when there are more 200,000 cases.

```python
covid[(covid.country == "US") & (covid.cases >= 200000) & (covid.type == "confirmed")]
```

### Practice problem 4:

Create a plot with the data from the previous problem (over 200,00 cases, confirmed COVID-19 in the US) where the y-axis is the number of confirmed cases and the x-axis is the date.

**Hints:**
- Make the axes labels descriptive
- Rotate the dates on the x-axis 90 degrees for legibility

```python
# subset data and save as data_subset
data_subset = covid[(covid.country == "US") & (covid.cases >= 200000) & (covid.type == "confirmed")]

# visualize with plotnine
(p9.ggplot(data=data_subset,
           mapping=p9.aes(x='date', y='cases'))
 + p9.geom_point()
 + p9.theme(axis_text_x=p9.element_text(angle = 90))
)
```

### Practice problem 6:

Find the maximum and minimum dates in this dataset.

```python
max = covid.date.max()
min = covid.date.min()

print("This dataset includes COVID data from ", min, "to", max)
```

### Practice problem 7:

Create a bar graph showing the total number of cases for each "type" (confirmed, death, recovered) using plotnine.

```python
(p9.ggplot(data=covid,
           mapping=p9.aes(x = 'type', y = "cases"))
     + p9.geom_col(alpha = 0.2, 
                   color = 'green')
)
```

### Practice problem 8:

Create a bar graph of the total number of confirmed cases for every country above the 100 degrees longitude using matplotlib.

```python
# filter to confirmed type only
covid_confirmed = covid[(covid.type == 'confirmed') & (covid.long > 100)]
covid_confirmed
```

### Practice problem 9:
Create a scatterplot of confirmed cases with the x-axis as the longitude and the y-axis as the latitude

```python
(p9.ggplot(covid_confirmed, 
           p9.aes(x = 'long',
                 y = 'lat'))
     + p9.geom_jitter()
)
```
